### Long vs Wide Data 

```python
#wide format: where we have a single row for every data point with multiple columns to hold the values of various attributes 
```

```python
#Long format: where, for each data point we have as many rows as the number of attriubutes and each row 
# contains the value of particular attribute for a given data point
```

```python
#melt -> wide to long 
```

```python
pd.DataFrame({'cse':[120]})
```

```python
pd.DataFrame({'cse':[120]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]})
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name='branch',value_name='num_students')
```

```python
x=pd.DataFrame(
        {
            'branch':['cse','ece','mec'],
            '2020':[100,150,60],
            '2021':[120,130,80],
            '2022':[150,140,70]
            
            
            
        }

)
```

```python
x
```

```python
x.melt()
```

```python
x.melt(id_vars=['branch'])
```

```python
x.melt(id_vars=['branch'],var_name='year',value_name='students')
```
